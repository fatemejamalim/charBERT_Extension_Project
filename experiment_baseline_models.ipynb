## Run MLM

BertCheckpoint: bert-base-cased

Dataset: Wiki_eng (resized)

Output Dir: nlp_charbert/model/output/general/MLM_cased/wiki_eng

```python
!python3 ./CharBERT/run_lm_finetuning.py \
    --model_type bert \
    --model_name_or_path bert-base-cased \
    --do_train \
    --do_eval \
    --train_data_file "./datasets/wikipedia_resized/wiki_eng_train.txt" \
    --eval_data_file  "./datasets/wikipedia_resized/wiki_eng_val.txt" \
    --term_vocab "./CharBERT/data/dict/term_vocab" \
    --learning_rate 5e-5 \
    --num_train_epochs 2 \
    --char_vocab "./CharBERT/data/dict/bert_char_vocab" \
    --mlm_probability 0.2 \
    --input_nraws 1000 \
    --save_steps 10000 \
    --block_size 384 \
    --mlm \
    --overwrite_output_dir \
    --output_dir  "./model/output/general/MLM_cased/wiki_eng"
```
